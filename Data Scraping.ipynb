# Creating the Dataset

I will be using data scraped from the twitter accounts of prominent national politicians using SNscrape. I will be getting all tweets from these polticians since Donald Trump became POTUS on 1/20/2017. I will be dividing this data into 4 subsets, Liberal Democrats, Moderate Democrats, Moderate Republicans, and Conservative Republicans. Then, I will create a dataset for each party, and then combine all of this data into one dataframe.

## Imports
